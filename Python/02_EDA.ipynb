```python
sales = pd.read_csv('sales_data_cleaned.csv')
sales.head()
```

```python
# Tổng revenue theo year và month
monthly_revenue = (sales.groupby(['year','month'])['revenue'].sum().reset_index())
monthly_revenue
```

```python
# Revenue theo region
sales.groupby('region')['revenue'].sum().reset_index()
```

```python
# Store-level Analysis
store_revenue=sales.groupby('store_id')['revenue'].sum().reset_index()
store_revenue
```

```python
top_5_store = store_revenue.sort_values(by='revenue', ascending=False).head(5)
top_5_store
```

```python
bottom_5_store = store_revenue.sort_values('revenue').head(5)
bottom_5_store
```
